```python
# Reload the data files
forecast_data = pd.read_csv('combined_forecasts_2005-2023.csv')
weather_data = pd.read_csv('C:/Users/madsh/OneDrive/Dokumenter/kandidat/FÃ¦llesmappe/Speciale/Data/Forbrug og Vejr Data.csv')

# Convert relevant columns to datetime format
forecast_data['valid_time'] = pd.to_datetime(forecast_data['valid_time'])
forecast_data['time'] = pd.to_datetime(forecast_data['time'])
weather_data['HourDK'] = pd.to_datetime(weather_data['HourDK'])

# Resample the weather data to match the forecast data
weather_data_resampled = weather_data.set_index('HourDK').resample('6H').mean().reset_index()

# Merge forecast and weather data based on the "valid_time" and "HourDK" columns
merged_data = pd.merge(forecast_data, weather_data_resampled, left_on='valid_time', right_on='HourDK', how='inner')

# Convert the 'step' column to a Pandas Timedelta object
merged_data['step'] = pd.to_timedelta(merged_data['step'])

# Calculate the errors for mean temperature
merged_data['error_mean'] = merged_data['t2m'] - merged_data['temp_mean_past1h']

# Select only relevant columns for clustering: 'step_hours', 't2m', 'error_mean'
merged_data['step_hours'] = merged_data['step'].dt.total_seconds() / 3600
cluster_data = merged_data[['step_hours', 't2m', 'error_mean']].copy()

# Apply linear interpolation to fill missing values in the 'error_mean' column
# The data is already resampled to 6-hour intervals, as specified earlier during the merging process
merged_data['error_mean'].interpolate(method='linear', inplace=True)

# Update the cluster_data DataFrame with the interpolated values
cluster_data['error_mean'] = merged_data['error_mean']

# Check again for missing values to confirm they have been handled
missing_values_updated = cluster_data.isnull().sum()
missing_values_updated

# Standardize the data
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)

# Determine the optimal number of clusters using the "elbow method"
inertia = []
for i in range(1, 51):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(cluster_data_scaled)
    inertia.append(kmeans.inertia_)

# Plot the inertia values to find the "elbow"
plt.figure(figsize=(10, 6))
plt.plot(range(1, 51), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
# Perform K-means clustering with 10 clusters
kmeans_10 = KMeans(n_clusters=10, random_state=0)
cluster_labels_10 = kmeans_10.fit_predict(cluster_data_scaled)

# Add the new cluster labels to the original DataFrame
merged_data['cluster_label_10'] = cluster_labels_10

# Explore the characteristics of each of the 10 clusters
cluster_summary_10 = merged_data.groupby('cluster_label_10').agg({
    'step_hours': ['min', 'max', 'mean'],
    't2m': ['min', 'max', 'mean'],
    'error_mean': ['min', 'max', 'mean', 'std']
}).reset_index()

cluster_summary_10.columns = ['Cluster', 'Min Step (hours)', 'Max Step (hours)', 'Mean Step (hours)',
                              'Min Temp (C)', 'Max Temp (C)', 'Mean Temp (C)',
                              'Min Error (C)', 'Max Error (C)', 'Mean Error (C)', 'Std Error (C)']

# Create a plot to visualize the clusters based on Mean Step and Mean Temperature
plt.figure(figsize=(12, 8))
plt.scatter(merged_data['step_hours'], merged_data['t2m'], c=merged_data['cluster_label_10'], cmap='viridis')
plt.colorbar(label='Cluster Label')
plt.xlabel('Step (hours)')
plt.ylabel('Temperature (C)')
plt.title('K-means Clustering with 10 Clusters')
plt.show()

cluster_summary_10
```
